```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=list(np.arange(0,32,4))     # np.arange(0,multiple of n+1,n+1)
x2=(np.array(x1)+1).tolist()
x3=(np.array(x2)+1).tolist()
```

```python
random_list=np.arange(0,8,1)
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')
#source : this function is taken from matplotlib documentation.
```

```python
fig, ax = plt.subplots(figsize=(16,8))
a = ax.bar(x1,random_list,label = "legend1")
b = ax.bar(x2,random_list,label = "legend2")
c = ax.bar(x3,random_list,label = "legend3")
plt.xticks(x2,["a","b","c","d","e","f","g","h"])
autolabel(a)
autolabel(b)
autolabel(c)
ax.legend()
plt.title("title")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
fig.tight_layout()
plt.grid()
```
